```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
from __future__ import division
from math import log
```

```python
# Entropy Calculation Using Python
def entropy(pi):
    '''
    return the Entropy of a probability distribution:
    entropy(p) =  SUM (Pi * log(Pi) )
    defintion:
            entropy is a metric to measure the uncertainty of a probability distribution.
    entropy ranges between 0 to 1
    Low entropy means the distribution varies (peaks and valleys).
    High entropy means the distribution is uniform.
    '''

    total = 0
    for p in pi:
        p = p / sum(pi)
        if p != 0:
            total += p * log(p, 2)
        else:
            total += 0
    total *= -1
    return total
```

```python
# Informatin Gain Calculation Using Python

def gain(d, a):
    '''
    return the information gain:
    gain(D, A) = entropy(D) SUM ( |Di| / |D| * entropy(Di) )
    '''

    total = 0
    for v in a:
       
        print("entropy of v",v,":-",entropy(v))
        total += sum(v) / sum(d) * entropy(v)
        print("*** total information gain",total)
    
    print("Total entropy of whole Tree [yes,no]-->",d,"<-- is -->",entropy(d))
    
    gain = entropy(d) - total
    return gain
```

```python
def gini_impurity (d,v):
    p_sum = 0
    gini_A=0
    for indsum in v:
      totalofindividuallabel=sum(indsum)
    for key in v:      
      p_sum = p_sum  +  (key[0]/sum(key)) * (key[0]/sum(key))+  (key[1]/sum(key)) * (key[1]/sum(key))
      p_sum=1-p_sum
      gini_A = gini_A+(totalofindividuallabel/sum(d)) *p_sum
    gini = gini_A
    return gini
```

```python
# set of example of the dataset
playTennis = [9, 5] # Yes, No

# attribute, number of members (feature)
outlook = [
    [4, 0],  # overcase
    [2, 3],  # sunny
    [3, 2]   # rain
]
temperature = [
    [2, 2],  # hot
    [3, 1],  # cool
    [4, 2]   # mild
]
humidity = [
    [3, 4],  # high
    [6, 1]   # normal
]
wind = [
    [6, 2],  # weak
    [3, 3]   # strong
]
print("************************************************************************************************************************************************")
print("Information Gain with Outlook:-",gain(playTennis, outlook))
#print("gini",totalginiImpurity(playTennis,outlook))
print("gini impurity -->",gini_impurity(playTennis,outlook),"<-- gini gain is -->",1-gini_impurity(playTennis,outlook))

print("***********************************************************************")
print("Information Gain with Temperature",gain(playTennis, temperature))
print("gini",gini_impurity(playTennis,temperature))
print("***********************************************************************")
print("Information Gain with Humidity",gain(playTennis, temperature))
print("gini",gini_impurity(playTennis,temperature))
print("***********************************************************************")
print("Information Gain with Wind",gain(playTennis, wind))
print("gini",gini_impurity(playTennis,wind))
print("**************************************************************************************************************************************************")
print("Since infromation gain of Outlook -->",gain(playTennis, outlook),"<-- is more compared to other features so outlook will be choosen as Root Node")
```
